```python
import pandas as pd
import numpy as np
import argparse
import os
from functools import partial

import pickle
import mlflow
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe
from hyperopt.pyll import scope

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report


def read_dataframe(data_path, filename):
    df = pd.read_csv(os.path.join(data_path, filename))
    return df

def preprocess(df):
    df["prognosis"] = df["prognosis"].apply(lambda x: x.replace("Paralysis (brain hemorrhageH", "Paralysis (brain hemorrhage)"))
    return df
    
```

```python
def train(params, X_train, y_train, X_test, y_test, df_train):
    model = RandomForestClassifier(**params, n_jobs=-1)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    score = f1_score(y_test, preds, average='macro')
    print("F1 Score: {}".format(score))
    target_names = df_train["prognosis"].unique()
    print(classification_report(y_test, preds, target_names=target_names))
    filename = 'rf_model.sav'
    pickle.dump(model, open(filename, 'wb'))

```

```python
print("Reading data files...")
data_path = os.path.join("..", "data", "disease-symptom-prediction")
df_train = read_dataframe(data_path, "Training.csv")
df_test = read_dataframe(data_path, "Testing.csv")

print("Preprocess data...")
df_train_processed = preprocess(df_train)
df_test_processed = preprocess(df_test)

print("Splitting into train and test set...")
X_train = df_train_processed.drop("prognosis", axis=1).values
y_train = df_train_processed["prognosis"].values
X_test = df_test_processed.drop("prognosis", axis=1).values
y_test = df_test_processed["prognosis"].values

print("Training model using best parameters...")
params = {
'max_depth': 13.995904464623607,
'n_estimators': 655,
'criterion': "gini",
'min_samples_leaf': 3
}

train(params, X_train, y_train, X_test, y_test, df_train)
```
